```python
import pandas as pd
```

```python
bikes_data_frame = pd.read_csv("london_merged.csv")
```

```python
bikes_data_frame.info()
```

```python
bikes_data_frame
```

```python
new_columns={
    'timestamp': 'time',
    'cnt': 'count',
    't1': 'temp_real_C',
    't2': 'temp_feels_like_C',
    'hum': 'humidity_percent',
    'wind_speed': 'wind_speed_kph',
    'weather_code': 'weather',
    'is_holiday': 'is_holiday',
    'is_weekend': 'is_weekend',
    'season': 'season'
}
bikes_data_frame.rename(new_columns, axis=1, inplace=True)
```

```python
bikes_data_frame
```

```python
bikes_data_frame['humidity_percent'] = bikes_data_frame['humidity_percent'] / 100
```

```python
bikes_data_frame
```

```python
season_code_map = {
    '0.0': 'spring',
    '1.0': 'summer',
    '2.0': 'autumn',
    '3.0': 'winter'
}

weather_code_map = {
    '1.0': 'Clear',
    '2.0': 'Scattered clouds',
    '3.0': 'Broken clouds',
    '4.0': 'Cloudy',
    '7.0': 'Rain',
    '10.0': 'Rain with thunderstorm',
    '26.0': 'Snowfall'
}

bikes_data_frame['season'] = bikes_data_frame['season'].astype(str)
bikes_data_frame['season'] = bikes_data_frame['season'].map(season_code_map)

bikes_data_frame['weather'] = bikes_data_frame['weather'].astype(str)
bikes_data_frame['weather'] = bikes_data_frame['weather'].map(weather_code_map)
```

```python
bikes_data_frame
```

```python
bikes_data_frame.to_excel("london_bikes_data.xlsx", sheet_name="Data")
```
